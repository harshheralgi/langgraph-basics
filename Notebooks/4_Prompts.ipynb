```python
import os
from dotenv import load_dotenv
load_dotenv()

if os.environ['GROQ_API_KEY']:
    print("Groq API Key Set!")
else:
    raise ValueError("Groq API Key is not set!!")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model = "llama-3.1-8b-instant")

llm.invoke("Hello!").content    
```

## **PROMPTS IN LANGGRAPH**

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant that helps to create posts for social media"),
        ("human","I want to create a post on {topic}")
    ]
)
```

```python
prompt.invoke({"topic":"Data Privacy in digital era."})
```

```python
chain = prompt | llm

chain.invoke({"topic":"Data Privacy in digital era."})
```
